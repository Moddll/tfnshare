# Mean Reversion Strategy

**Signal** :  
        - crossover :  ma(5) > ma(20) and last.ma(5) < last.ma(20)
        - crossbelow:  ma(5) < ma(20) and last.ma(5) > last.ma(20)

## Get Stock Data

### Initial Database

```python
import sys
sys.path.insert(0, '../stock')

import pandas as pd
import numpy as np

from tfnstock import RwDatabase
from tfnstock import FormatStockData
# set database
db = '../findata/cse.db'
dbc = RwDatabase(db)
```

### Get Stock Data

```python
# Get stock data from database
symbol = "GET"
data = pd.DataFrame()
formatdata = FormatStockData(db)

data = formatdata.ma_stockdata(symbol)
data.tail(5)
```
